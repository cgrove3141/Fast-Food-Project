```python
# Dependencies
import pandas as pd
```

```python
# Name of the CSV file
file = 'data/fastfood.csv'
```

```python
# The correct encoding must be used to read the CSV in pandas
df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
# Preview of the DataFrame
df.head()
```

```python
# Drop all rows with missing information
df = df.dropna(how='any')
```

```python
# Verify dropped rows
df.count()
```

```python
dfObj = pd.DataFrame(df, columns=['name', 'address', 'latitude', 'longitude'])
duplicateDFRow = dfObj[dfObj.duplicated()]
duplicateDFRow
```

```python
# Use pd.to_numeric() method to convert the datatype of the Amount column
result_df=df.drop_duplicates(subset=['latitude', 'longitude'], keep='last')
result_df
```

```python
# Verify dropped rows
result_df.count()
```

```python
result_df.to_csv("data/FastFoodCleaned.csv", index=False)
```
